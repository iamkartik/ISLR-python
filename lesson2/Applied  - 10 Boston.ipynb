```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# loading boston dataset from sklearn
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
boston.keys()
```

```python
print(boston.DESCR)
```

```python
boston.target
```

```python
housing = pd.DataFrame(data=boston['data'],columns=boston['feature_names'])
```

```python
housing.head()
```

```python
housing['MEDV'] = boston.target
```

```python
housing.info()
```

```python
sns.set()

colsize = len(housing.columns)

firstcols = housing.columns[:colsize//2]
lastcols = housing.columns[colsize//2:]

sns.pairplot(housing[firstcols])
```

```python
sns.pairplot(housing[lastcols])
```

```python
sns.pairplot(housing[['CRIM','NOX','RM','AGE','DIS','B','LSTAT']])
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(housing.corr(),annot=True)
```

```python
housing.CRIM.plot(kind='hist')
```

```python
fig,(ax1,ax2,ax3) = plt.subplots(ncols=3,figsize=(12,6))
sns.distplot(housing['CRIM'],bins=30,ax=ax1)
sns.distplot(housing['TAX'],bins=30,ax=ax2)
sns.distplot(housing['PTRATIO'],bins=30,ax=ax3)
plt.show()
```
