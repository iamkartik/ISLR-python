```python
import warnings 
warnings.filterwarnings('ignore')
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
college = pd.read_csv('../datasets/College.csv',index_col=0)
```

```python
college.head()
```

```python
college.reset_index(inplace=True)
college.rename(columns={'index':'name'},inplace=True)
```

```python
college.describe()
```

```python
sns.pairplot(college)
```

```python
sns.boxplot(x='Private',y='Outstate',data=college)
```

```python
college['elite'] = college['Top10perc'].apply(lambda x:'Yes' if x>50 else 'No')
```

```python
college.head(5)
```

```python
sns.boxplot(x='elite',y='Outstate',data=college)
```

```python
sns.set()
plt.figure(figsize=(15,12))
fig , ax = plt.subplots(ncols=2, nrows=2)
sns.distplot(college['Room.Board'],bins=30,ax=ax[0,0])
sns.distplot(college['Personal'],bins=30,ax=ax[0,1])
sns.distplot(college['Books'],bins=30,ax=ax[1,0])
sns.distplot(college['PhD'],bins=30,ax=ax[1,1])
plt.tight_layout()
fig.show()
```
